# Spotify Dataset Analysis using Python

# Objective

The objective of the Spotify dataset analysis is to gain valuable insights and actionable recommendations by leveraging the extensive collection of data from the popular music streaming platform. By analyzing various aspects of the dataset, including user preferences, music genres, artist popularity, and listening trends, the goal is to enhance user experience, optimize content curation, and support strategic decision-making for Spotify and its stakeholders.

## 01. Importing python Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

## 02. Loading Dataset  using pandas

```python
df_track=pd.read_csv('tracks.csv')
df_track.head()
```

## 03. Checking null values in different columns

```python
df_track.isnull().sum()
```

## 04.  Checking information about every columns using "info" method

```python
df_track.info()
```

## 05. Sorting the values on the basis of popularity and getting least popular songs

```python
sorted_df=df_track.sort_values('popularity',ascending=True).head(10)
sorted_df
```

## 06. Generating the statical description of the defferent columns using "DESCRIBE" method 

```python
df_track.describe().transpose()
```

## 07. Extracting the values which popularity is greater than 90

```python
most_popular=df_track.query('popularity>90',inplace=False).sort_values('popularity',ascending=True).tail(10)
most_popular
```

## 08.Setting index to the "release_date" columns

```python
df_track.set_index("release_date",inplace=True)
df_track.index=pd.to_datetime(df_track.index)
df_track.head()
```

## 09.Getting particular value using "iloc" method 

```python
df_track[["artists"]].iloc[25]
```

## 10. Converting duration from milliseconds to seconds

```python
df_track["duration"]=df_track["duration_ms"].apply(lambda x: round(x/1000))
df_track.drop("duration_ms",inplace=True ,axis=1)
```

```python
df_track.duration.head()
```

## 11. Creating correlation map using seaborn heatmap method 

```python
corr_df=df_track.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("Correlation Heatmap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

## 12.Creating a sample data which takes 0.4 percent of the total data 

```python
sample_df=df_track.sample(int(0.004*len(df_track)))
```

```python
print(len(sample_df))
```

## 13. Creating regration plot  between loudness and energy using seaborn
 

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness",x="energy",color="c").set(title="loudness vs energy correlation")
```

## 14. Creating regration plot  between popularity  and acousticness using seaborn


```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="popularity",x="acousticness",color="b").set(title="popularity vs acousticness correlation")
```

## 15. Creating a new column from release_date column

```python
df_track['dates']=df_track.index.get_level_values('release_date')
df_track.dates=pd.to_datetime(df_track.dates)
years=df_track.dates.dt.year
```

## 16. Creating Distribution plot to see total number of songs available on spotify per year 

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="NUMBER OF SONGs PER YEAR")
```

## 17. Creating bar plot between duration and years using seaborn.

```python
total_dr=df_track.duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title="year vs duration")
plt.xticks(rotation=90)
```

## 18. Loading another dataset 

```python
df_genre=pd.read_csv('SpotifyFeatures.csv')
```

```python
df_genre.head()
```

## 19. Creating bar plot for  duration of the songs in  different genres

```python
plt.title("Duration of the songs in different genre")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y="genre",x="duration_ms",data=df_genre)
plt.xlabel("duration in milli seconds")
plt.ylabel("genres")
```

## 20. Getting top 5 genre by popularity.

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous=df_genre.sort_values("popularity",ascending=False).head(10)
sns.barplot(y='genre',x='popularity',data=famous).set(title="top 5 genre by popularity")

```
